```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

## Data Source and Contents

```python
#import data
data = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')
```

```python
display(data.head())
display(data.describe())
display(data.info())
display(data.isnull().sum())
```

```python
#drop 'label' column due to instruction
data.drop(columns = 'Label', inplace=True)
```

```python
#display the dataset after drop down the column
display(data.head())
display(data.columns)
```

```python
#feature selection
#split the dataset in features varibales and target variables
feature_cols = ['IMDB Rating', 'Biography', 'Drama', 'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']
X = data[feature_cols]# Feature variables
Y = data[['Movie Name']] #target variables
```

```python
#train the model using the dataset
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X, Y)
```

```python
#IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No, Mystery = No, History = Yes
#substitute the data of 'the post' to calculate the distance to other movies
Target = classifier.kneighbors([[7.2, 1, 1, 0, 0, 0, 0, 1]], return_distance= False)
```

```python
#show the results
for item in Target:
    rec = data.iloc[item]['Movie Name']
   
print(f'Here are movies you might like based on your interest in "the post": \n{rec}')
```
